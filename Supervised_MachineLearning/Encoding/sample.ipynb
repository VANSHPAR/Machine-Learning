```python
from sklearn.preprocessing import LabelEncoder
import pandas as pd

df=pd.read_csv(r"C:\Users\Vansh Parmar\OneDrive\Attachments\Desktop\min\coding\Python\ML\Supervised_MachineLearning\Encoding\sample.csv")
# print(df)

df_label=df.copy()


#Label Encoding
le=LabelEncoder()
df_label['Gender_Encoded']=le.fit_transform(df_label['Gender'])
df_label['Passed_Encoded']=le.fit_transform(df_label['Passed'])

print('\nEncoded Data')
print(df_label[['Name','Gender','Gender_Encoded','Passed','Passed_Encoded']])

#One Hot Encoding
df_encoded=pd.get_dummies(df_label,columns=['City'])
df_encoded['City_Bangalore']=df_encoded['City_Bangalore'].astype(int)
df_encoded['City_Chennai']=df_encoded['City_Chennai'].astype(int)
df_encoded['City_Delhi']=df_encoded['City_Delhi'].astype(int)
df_encoded['City_Mumbai']=df_encoded['City_Mumbai'].astype(int)


print('\nOne Hot Encoded Data City')
print(df_encoded)

```

```python
import pandas as pd
from sklearn.preprocessing import StandardScaler,MinMaxScaler
from sklearn.model_selection import train_test_split

data={
    'Studyhours':[1,2,3,4,5],
    'Testscore':[40,50,60,70,80]
}
df=pd.DataFrame(data)

standard_scalar=StandardScaler()
standard_scaled=standard_scalar.fit_transform(df)

print("Standard Scaler Output")
print(pd.DataFrame(standard_scaled,columns=['Studyhours','Testscore']))

minmax_scalar=MinMaxScaler()
minmax_scaled=minmax_scalar.fit_transform(df)

print('Minmax Scalar output')
print(pd.DataFrame(minmax_scaled,columns=['Studyhours','Testscore']))

x=df[['Studyhours']]
y=df[['Testscore']]

x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=42)
print('Training data')
print(x_train)

print('Test data')
print(x_test)

print('Training data')
print(y_train)

print('Test data')
print(y_test)

#0.2 means 80% data for training pupose and 20% data for testing purpose

#    Studyhours
# 4           5
# 2           3
# 0           1
# 3           4
# index      Value
# Test data
#    Studyhours
# 1           2
```
